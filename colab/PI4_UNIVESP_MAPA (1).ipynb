Importando bibliotecas 


```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python


# Ler o arquivo CSV
df = pd.read_csv('https://raw.githubusercontent.com/Univesp-PI-2022/projeto-integrador-4/main/dados/censo1.csv',sep=";")

# Exibir o DataFrame resultante
df.head()
```

```python
df.count()
```

### Mostrando os lugares ###

```python
df[['Latitude', 'Longitude']].head()
```

```python
df['Latitude'] = df['Latitude'].apply(lambda x: float(str(x).replace(',', '.')))
df['Longitude'] = df['Longitude'].apply(lambda x: float(str(x).replace(',', '.')))
```

```python
df[['Latitude', 'Longitude']].head()
```

```python
df['Latitude'].count()
```

```python
df['Longitude'].count()
```

### Removendo Nan ###

```python
df.dropna(subset=['Latitude'], inplace=True)
```

```python
df.dropna(subset=['Longitude'], inplace=True)
```

```python
df.count()
```

```python
df['Latitude'].isna().sum()
```

```python
df['Longitude'].isna().sum()
```

```python
df = df.sample(n=100, random_state=42)
```

```python
df.count()
```

```python
mapa = folium.Map(location=[-23,-46], zoom_start=13)
```

```python
# Adicionar marcadores para cada lugar no dataset
for index, row in df.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']]).add_to(mapa)
```
